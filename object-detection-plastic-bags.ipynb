```python
zip_file_paths = ['train_images.zip', 'validation_images.zip']
```

```python
for path in zip_file_paths:
    file_name = str(path)
    with zipfile.ZipFile(file_name, 'r') as zip_ref:
        path = f'unzipped/{file_name.split(".")[0]}'
        zip_ref.extractall(path)
```

```python
def count_images(imgs_path):
    glob_imgs_path = glob.glob(imgs_path)
    counter = 0

    for _ in glob_imgs_path:
        counter += 1
    return counter


(
    count_images(imgs_path=data_path_validation),
    count_images(imgs_path=data_path_train),
)
```
